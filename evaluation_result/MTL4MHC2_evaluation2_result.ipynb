```python



bi_lstm = np.array(com_HLA3tre["Bi-lstm"])
task_lstm = np.array(com_HLA3tre["Multi-task learning"])
multi_head = np.array(com_HLA3tre["Multi-head attention"])


b_t_t_test = stats.ttest_rel(bi_lstm, task_lstm)
b_m_t_test = stats.ttest_rel(bi_lstm, multi_head)
t_m_t_test = stats.ttest_rel(task_lstm, multi_head)

print(b_t_t_test)
print(b_m_t_test)
print(t_m_t_test)
```

```python
bi_lstm_var = np.var(bi_lstm, ddof=1)  # unbiased estimate of variance of bi_lstm
task_lstm_var = np.var(task_lstm, ddof=1)  # unbiased estimate of variance of task_lstm
multi_head_var = np.var(multi_head, ddof=1)  # unbiased estimate of variance of multi_head

bi_lstm_df = len(bi_lstm) - 1  # degree of freedom of bi_lstm
task_lstm_df = len(task_lstm) - 1  # degree of freedom of task_lstm
multi_head_df = len(multi_head) - 1  # degree of freedom of multi_head

# F rate
f1 = bi_lstm_var / task_lstm_var
f2 = bi_lstm_var / multi_head_var
f3 = task_lstm_var / multi_head_var

#bi_lstm_var vs task_lstm_var
one_sided_pval1a = stats.f.cdf(f1, bi_lstm_df, task_lstm_df)  # p-value of one-side test
one_sided_pval2a = stats.f.sf(f1, bi_lstm_df, task_lstm_df)   # p-value of one-side test
two_sided_pvala = min(one_sided_pval1a, one_sided_pval2a) * 2  # p-value of two-sided test

print('bi_lstm_var vs task_lstm_var')
print('F:       ', round(f1, 3))
print('p-value: ', round(two_sided_pvala, 4))


#bi_lstm_var vs multi_head_var
one_sided_pval1b = stats.f.cdf(f2, bi_lstm_df, multi_head_df) 
one_sided_pval2b = stats.f.sf(f2, bi_lstm_df, multi_head_df)   
two_sided_pvalb = min(one_sided_pval1b, one_sided_pval2b) * 2  

print('bi_lstm_var vs multi_head_var')
print('F:       ', round(f2, 3))
print('p-value: ', round(two_sided_pvalb, 4))


#task_lstm_var vs multi_head_var
one_sided_pval1c = stats.f.cdf(f3, task_lstm_df, multi_head_df) 
one_sided_pval2c = stats.f.sf(f3, task_lstm_df, multi_head_df)   
two_sided_pvalc = min(one_sided_pval1c, one_sided_pval2c) * 2  

print('task_lstm_var vs multi_head_var')
print('F:       ', round(f3, 3))
print('p-value: ', round(two_sided_pvalc, 4))
```
